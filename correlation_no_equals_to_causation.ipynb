```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Constants
phi = 0.8
sigma = 1.0
n = 100

# Generate x0 as AR(1)
np.random.seed(42)
epsilon = np.random.normal(0, sigma, n)
x0 = np.zeros(n)
for t in range(1, n):
    x0[t] = phi * x0[t-1] + epsilon[t]

# Generate x1 and x2
eta1 = np.random.normal(0, sigma, n)
eta2 = np.random.normal(0, sigma, n)
x1 = x0 + eta1
x2 = x0 + eta2

# Plot time series
plt.figure(figsize=(4, 3), dpi=900)
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.xlabel('Time (t)')
plt.ylabel('Value')
plt.title('Time Series of x1 and x2')
plt.legend()
plt.show()

# Scatter plot x1 vs x2
slope, intercept, r_value, _, _ = linregress(x1, x2)
line_fit = slope * x1 + intercept
plt.figure(figsize=(4, 3), dpi=900)
plt.scatter(x1, x2, alpha=0.5)
plt.plot(x1, line_fit, 'r', label=f'Fit line: $R={r_value**2:.2f}$')
plt.xlabel('x$_1$')
plt.ylabel('x$_2$')
plt.title('x1 VS x2')
plt.legend()
plt.show()

# Scatter plot for eta1 vs eta2
slope_eta, intercept_eta, r_value_eta, _, _ = linregress(eta1, eta2)
line_fit_eta = slope_eta * eta1 + intercept_eta
plt.figure(figsize=(4, 3), dpi=900)
plt.scatter(eta1, eta2, alpha=0.5, color='green')
plt.plot(eta1, line_fit_eta, 'r', label=f'Fit line: $R={r_value_eta**2:.2f}$')
plt.xlabel('Noise in x$_1$ ($_1$)')
plt.ylabel('Noise in x$_2$ ($_2$)')
plt.title('Residual of x1 VS Residual of x2')
plt.legend()
plt.show()
```
